```python
#Bivariat Analysis
plt.title('Salary vs Experience')
plt.xlabel('Year of Experience')
plt.ylabel('Salary')
plt.scatter(df['YearExperince'],df['Salary'], color='lightcoral')
plt.show()
```

```python
#Ploting with seaborn
sns.scatterplot(x='YearExperince', y='Salary', data=df)
plt.title('Salary vs Experience')
plt.xlabel('Year of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
#Check corelation between 2 variable (if output >0 called positive and if output <0 called nagative)
correlation = df['YearExperince'].corr(df['Salary'])
print(correlation)
```
